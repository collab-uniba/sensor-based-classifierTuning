```python
import pandas as pd
import matplotlib.pyplot as plt

# Leggi il file CSV e crea il DataFrame
df = pd.read_csv("all.csv")

# Estrai il soggetto principale
df['main_subject'] = df['id'].str.extract(r'(S\d+)')

# Identifica i soggetti TOSEM e ICSE
tosem_subjects = [f'S{i}' for i in range(1, 11)]
icse_subjects = [f'S{i}' for i in range(11, 33)]

# Filtra i soggetti esistenti nel DataFrame
existing_tosem_subjects = [subject for subject in tosem_subjects if subject in df['main_subject'].unique()]
existing_icse_subjects = [subject for subject in icse_subjects if subject in df['main_subject'].unique()]

# Calcola il numero di istanze per TOSEM, ICSE e complessivamente
num_tosem_instances = df[df['main_subject'].isin(existing_tosem_subjects)].shape[0]
num_icse_instances = df[df['main_subject'].isin(existing_icse_subjects)].shape[0]
total_instances = df.shape[0]

# Crea una tabella di contingenza utilizzando il soggetto principale
contingency_table = pd.crosstab(df['main_subject'], df['valence'])

# Separa le conte per TOSEM e ICSE
tosem_counts = contingency_table.loc[existing_tosem_subjects].sum()
icse_counts = contingency_table.loc[existing_icse_subjects].sum()
all_counts = contingency_table.sum()

# Crea il DataFrame per la visualizzazione
result_df = pd.DataFrame({
    'ICSE': icse_counts,
    'TOSEM': tosem_counts,
    'ALL': all_counts
}).reindex(['positive', 'neutral', 'negative'])

# Aggiungi il numero di istanze
result_df.loc['Num Instances'] = [num_icse_instances, num_tosem_instances, total_instances]

# Plot della tabella
fig, ax = plt.subplots()
ax.axis('tight')
ax.axis('off')
table = ax.table(cellText=result_df.values,
                 colLabels=result_df.columns,
                 rowLabels=result_df.index,
                 cellLoc='center',
                 loc='center')

# Visualizza la tabella
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np

# Leggi il file CSV e crea il DataFrame
df = pd.read_csv("all.csv")

# Estrai il soggetto principale
df['main_subject'] = df['id'].str.extract(r'(S\d+)')

# Calcola il numero totale di istanze per ciascun soggetto
total_instances_per_subject = df['main_subject'].value_counts()

# Calcola il numero di istanze positive, negative e neutre per ciascun soggetto
subject_counts = df.groupby(['main_subject', 'valence']).size().unstack(fill_value=0)

# Calcola le percentuali
percentages = subject_counts.div(total_instances_per_subject, axis=0) * 100

# Ordina le righe in base ai soggetti TOSEM e ICSE
tosem_subjects = [f'S{i}' for i in range(1, 11)]
icse_subjects = [f'S{i}' for i in range(11, 33)]

ordered_subjects = tosem_subjects + icse_subjects

# Seleziona solo i soggetti presenti nel DataFrame
ordered_subjects = [subject for subject in ordered_subjects if subject in percentages.index]

# Seleziona solo le etichette desiderate
labels = ['positive', 'neutral', 'negative']
percentages = percentages.loc[ordered_subjects, labels]

# Prepara i dati per la heatmap
data = percentages.values

# Crea la heatmap utilizzando Matplotlib
fig, ax = plt.subplots(figsize=(4, 22))  # Imposta le dimensioni della figura

# Definisci la colormap
cmap = plt.cm.get_cmap('Reds')

# Crea la heatmap
heatmap = ax.imshow(data, cmap=cmap, interpolation='nearest')

# Aggiungi la barra dei colori
cbar = plt.colorbar(heatmap, ax=ax)
cbar.ax.tick_params(labelsize=14)  # Imposta la dimensione del testo della barra dei colori

# Etichette sugli assi
ax.set_xticks(np.arange(len(labels)))
ax.set_yticks(np.arange(len(percentages.index)))
ax.set_xticklabels(labels, fontsize=14)
ax.set_yticklabels(percentages.index, fontsize=14)

# Rotazione delle etichette sull'asse x
plt.xticks(rotation=45)

# Aggiungi le annotazioni con i valori
for i in range(len(percentages.index)):
    for j in range(len(labels)):
        text = ax.text(j, i, f'{data[i, j]:.1f}%', ha='center', va='center', color='black', fontsize=14)

# Titolo e label degli assi
plt.title('Percentuale di istanze per soggetto e etichetta', fontsize=18)
plt.xlabel('Etichetta', fontsize=16)
plt.ylabel('Soggetto', fontsize=16)

plt.tight_layout()
plt.show()
```
