```python
from sklearn.cluster import KMeans
import numpy as np
import pandas as pd
import csv
import os
import matplotlib.pyplot as plt
from sklearn.decomposition import PCA 
```

```python
#Read baselineFeatures.csv or baselineFeatures_no_S5_S8.csv
with open('C:/Users/Utente/Desktop/sensor-based-classifierTuning/AlbertaProgetto/baselineFeatures.csv', 'r') as f:  
#with open('C:/Users/Utente/Desktop/sensor-based-classifierTuning/AlbertaProgetto/baselineFeatures_no_S5_S8.csv', 'r') as f:          

    reader = csv.reader(f, delimiter='t')
    head = ''
    lines = []
    #For each subject
    for i, line in enumerate(reader):
        #Save header
        if i == 0:
            head = line
        else:
            #Create the testing file
            with open('C:/Users/Utente/Desktop/sensor-based-classifierTuning/AlbertaProgetto/LOSO_Output_testing/'+ str(i) + '.csv', 'w', newline='') as file:
                writer = csv.writer(file)
                writer.writerow(head)
                writer.writerow(line)
            file.close()
        
        #Save the data of all subjects
        lines.append(line)    

    
    to_write = []
    #For each subject
    for i in range(len(lines)):
        to_write = lines.copy()
        if i > 0:
            #Remove the subject who is in testing file
            to_write.remove(lines[i])
            #Create the training file
            with open('C:/Users/Utente/Desktop/sensor-based-classifierTuning/AlbertaProgetto/LOSO_Output_training/'+ str(i) + '.csv', 'w', newline='') as file:
                    writer = csv.writer(file)
                    for line in to_write:
                        writer.writerow(line)
            file.close()
        
    f.close      
```

```python
#Find the ids of who is in cluster
def who_in_cluster(cluster, dataset):
    ids = []
    if len(cluster) == 0: return ids
    for cluster_data in cluster:
        for data in dataset:
            for i in range(0, 12):
                if cluster_data[i] != data[i+1]:
                    break
                else:
                    if i == 11:
                        ids.append(data[0])
    return ids

#Write on file the ids
with open('C:/Users/Utente/Desktop/sensor-based-classifierTuning/AlbertaProgetto/ids.csv', 'w', newline='') as file:
    
    #For each subject
    for i in range(1, 23):
        
    #For each subject without S5 and S8
    #for i in range(1,21):
    
        dataset = np.empty([0, 13])
        
        #take the training file
        with open ('C:/Users/Utente/Desktop/sensor-based-classifierTuning/AlbertaProgetto/LOSO_Output_training/'+ str(i) + '.csv', 'r') as f:
            reader_training = csv.reader(f, delimiter='t')
            datas = np.empty([0, 12])
            
            for j, subject_training in enumerate(reader_training):
                #Ignore header
                if j > 0: 
                    #for each line of the training file
                    for z, line in enumerate(subject_training):
                        #split the datas
                        dat = [list(map(float, line.split(',')))]
                        #remove id
                        dat_no_idx = np.delete(dat, 0, 1)
                        #create a dataset with id
                        dataset = np.append(dataset, dat, axis=0)
                        #create a dataset without id
                        datas = np.append(datas, dat_no_idx, axis=0)

            #K-means on dataset without ids with K=2
            kmeans =KMeans(n_clusters=2, random_state=0, n_init="auto").fit(datas)
            
            #PCA create a two-dimensional view of n features
            pca = PCA(n_components=2)
            datas_2d = pca.fit_transform(datas)
            print('CENTERS \n',kmeans.cluster_centers_)
            cluster_1_indices = np.where(kmeans.labels_ == 0)[0]
            cluster_2_indices = np.where(kmeans.labels_ == 1)[0]
            #cluster_3_indices = np.where(kmeans.labels_ == 2)[0]
            #cluster_4_indices = np.where(kmeans.labels_ == 3)[0]

            print("Dati nel Cluster 1:")
            print(datas[cluster_1_indices])

            print("\nDati nel Cluster 2:")
            print(datas[cluster_2_indices])
            
            #print("\nDati nel Cluster 3:")
            #print(datas[cluster_3_indices])
            
            #print("\nDati nel Cluster 4:")
            #print(datas[cluster_4_indices])
            
            plt.scatter(datas_2d[:, 0], datas_2d[:, 1], c=kmeans.labels_, cmap='viridis',label='Data Points')
            plt.scatter(kmeans.cluster_centers_[:, 0], kmeans.cluster_centers_[:, 1], s=300, c='red', alpha=0.5, label='Centroids')  
            plt.legend()
            plt.title('Training Data - Cluster Visualization')
            plt.show() 


        #Take the testing files
        with open ('C:/Users/Utente/Desktop/sensor-based-classifierTuning/AlbertaProgetto/LOSO_Output_testing/'+ str(i) + '.csv', 'r') as f1:
            reader_testing = csv.reader(f1, delimiter='t')
            datas_testing = np.empty([0, 12])
            ids = []
            for x, subject_testing in enumerate(reader_testing):
                #Ignore header
                if x > 0: 
                    #Split the datas
                    for t, line in enumerate(subject_testing):
                        splitted = list(map(float, line.split(',')))
                        #Remove id
                        dat_test = np.delete([splitted], 0, 1)
                        #Create dataset without ids
                        datas_testing = np.append(datas_testing,dat_test, axis=0)
 
            #Predict on testing subject
            pred = kmeans.predict(datas_testing)
            print('s'+ str(i) + ' appartiene al cluster: ', pred)

            #If the predicted cluster is 1
            if pred[0] == 0:
                #find the others ids in this cluster
                ids = who_in_cluster(datas[cluster_1_indices], dataset)
            #If the predicted cluster is 2
            elif pred[0] == 1:
                #find the others ids in this cluster
                ids = who_in_cluster(datas[cluster_2_indices], dataset)
            #elif pred[0] == 2:
                #ids = who_in_cluster(datas[cluster_3_indices], dataset)
            #elif pred[0] == 3:
                #ids = who_in_cluster(datas[cluster_4_indices], dataset)
            
            print("Di cui fanno parte anche: ", ids)
            
            #Add the id of testing subject
            ids.append(str(i)+".1")
            
            #Write all ids on file
            writer = csv.writer(file)
            writer.writerow(ids)

            datas_testing_2d = pca.transform(datas_testing)
            plt.scatter(datas_testing_2d[:, 0], datas_testing_2d[:, 1], c=pred, cmap='viridis', label='Data Point')
            plt.scatter(kmeans.cluster_centers_[:, 0], kmeans.cluster_centers_[:, 1], s=300, c='red', alpha=0.5, label='Centroids')  
            plt.legend()
            plt.title('Testing Data - Cluster Visualization)')
            plt.show()
            
file.close()
```
